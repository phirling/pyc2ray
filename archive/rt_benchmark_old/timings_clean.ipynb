```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 12
save = False

plt.rcParams.update({"text.usetex": True,'font.family': 'serif','font.size':10})
```

# Test A: Constant $N_{src}$, varying $r_{RT}$

```python
r_RT = np.linspace(10,200,20)
res_octa_10src = np.loadtxt("timings_clean/t_vs_r/result_octa_10src.out")
res_octa_100src = np.loadtxt("timings_clean/t_vs_r/result_octa_100src.out")
res_octa_1000src = np.loadtxt("timings_clean/t_vs_r/result_octa_1000src.out")

res_no_octa_10src = np.loadtxt("timings_clean/t_vs_r/result_no_octa_10src.out")
res_no_octa_100src = np.loadtxt("timings_clean/t_vs_r/result_no_octa_100src.out")
res_no_octa_1000src = np.loadtxt("timings_clean/t_vs_r/result_no_octa_1000src.out")
```

```python
t_octa_10src = res_octa_10src[:,1]
t_octa_100src = res_octa_100src[:,1]
t_octa_1000src = res_octa_1000src[:,1]

t_no_octa_10src = res_no_octa_10src[:,1]
t_no_octa_100src = res_no_octa_100src[:,1]
t_no_octa_1000src = res_no_octa_1000src[:,1]
```

```python
plt.figure(figsize=(5,5))
plt.plot(r_RT,t_no_octa_10src/t_octa_10src,'o--',color='coral',label="$N_{src}=10$")
plt.plot(r_RT,t_no_octa_100src/t_octa_100src,'s-',color='crimson',label="$N_{src}=100$")
plt.plot(r_RT,t_no_octa_1000src/t_octa_1000src,'^-.',color='black',label="$N_{src}=1000$")
plt.xlabel("$r_{RT}$ [Grid Coords.]")
plt.ylabel("$t_{S} / t_{A}$")
plt.xlim(0,210)
plt.legend(fontsize=14)
if save: plt.savefig("t_vs_r.png",dpi=300,bbox_inches='tight')
```

# Test B: Constant $r_{RT}$, varying $N_{src}$

```python
numsrc = [1, 10, 50, 100, 300, 500, 750, 1000, 1500]
res_octa_r50 = np.loadtxt("timings_clean/t_vs_numsrc/result_octa_r50.out")
res_octa_r100 = np.loadtxt("timings_clean/t_vs_numsrc/result_octa_r100.out")
res_octa_r200 = np.loadtxt("timings_clean/t_vs_numsrc/result_octa_r200.out")

res_no_octa_r50 = np.loadtxt("timings_clean/t_vs_numsrc/result_no_octa_r50.out")
res_no_octa_r100 = np.loadtxt("timings_clean/t_vs_numsrc/result_no_octa_r100.out")
res_no_octa_r200 = np.loadtxt("timings_clean/t_vs_numsrc/result_no_octa_r200.out")
```

```python
t_octa_r50 = res_octa_r50[:,1]
t_octa_r100 = res_octa_r100[:,1]
t_octa_r200 = res_octa_r200[:,1]

t_no_octa_r50 = res_no_octa_r50[:,1]
t_no_octa_r100 = res_no_octa_r100[:,1]
t_no_octa_r200 = res_no_octa_r200[:,1]
```

```python
plt.figure(figsize=(5,5))
plt.plot(numsrc,t_no_octa_r50/t_octa_r50,'o--',color='coral',label="$r_{RT}=50$")
plt.plot(numsrc,t_no_octa_r100/t_octa_r100,'s-',color='crimson',label="$r_{RT}=100$")
plt.plot(numsrc,t_no_octa_r200/t_octa_r200,'^-.',color='black',label="$r_{RT}=200$")
plt.xlabel("$N_{src}$")
plt.ylabel("$t_{S} / t_{A}$")
plt.legend(fontsize=14)
if save: plt.savefig("t_vs_numsrc.png",dpi=300,bbox_inches='tight')
```

```python
fig, ax = plt.subplots(1,2,figsize=(8,4))

ax[0].set_box_aspect(1)
ax[1].set_box_aspect(1)

ms = 5
c1 = 'cornflowerblue'
c2 = 'darksalmon'
c3 = 'black'

ax[0].plot(r_RT,t_no_octa_10src/t_octa_10src,'o--',ms=ms,color=c1,label="$N_{src}=10$")
ax[0].plot(r_RT,t_no_octa_100src/t_octa_100src,'s-',ms=ms,color=c2,label="$N_{src}=100$")
ax[0].plot(r_RT,t_no_octa_1000src/t_octa_1000src,'^-.',ms=ms+1,color=c3,label="$N_{src}=1000$")
ax[0].set_xlabel("$r_{RT}$ [Grid Cells]")
ax[0].set_ylabel("$t_{S} / t_{A}$")
ax[0].legend()

ax[1].plot(numsrc,t_no_octa_r50/t_octa_r50,'o--',ms=ms,color=c1,label="$r_{RT}=50$")
ax[1].plot(numsrc,t_no_octa_r100/t_octa_r100,'s-',ms=ms,color=c2,label="$r_{RT}=100$")
ax[1].plot(numsrc,t_no_octa_r200/t_octa_r200,'^-.',ms=ms+1,color=c3,label="$r_{RT}=200$")
ax[1].set_xlabel("$N_{src}$")
ax[1].set_ylabel("$t_{S} / t_{A}$")
ax[1].legend()
```
